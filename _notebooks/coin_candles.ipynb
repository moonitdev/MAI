```python
import requests

# https://api.upbit.com/v1/candles/minutes/unit
# unit: 분 단위. 가능한 값 : 1, 3, 5, 15, 10, 30, 60, 240
# 일 단위: https://api.upbit.com/v1/candles/days
# 주 단위: https://api.upbit.com/v1/candles/weeks
# 월 단위: https://api.upbit.com/v1/candles/months
url = "https://api.upbit.com/v1/candles/minutes/3"

# count 200개까지 요청 가능
querystring = {"market":"KRW-BTC","count":"10", "to":"2020-01-31 09:58:00"}

response = requests.request("GET", url, params=querystring)

print(response.text)
```

```python
# https://apidocs.bithumb.com/docs/candlestick

import requests

# {order_currency} = 주문 통화(코인), 기본값 : BTC
# {payment_currency} = 결제 통화(마켓), 기본값 : KRW
# {chart_intervals} = 차트 간격, 기본값 : 24h {1m, 3m, 5m, 10m, 30m, 1h, 6h, 12h, 24h 사용 가능}
url = "https://api.bithumb.com/public/candlestick/{order_currency}_{payment_currency}/{chart_intervals}"
# url = "https://api.bithumb.com/public/candlestick/{}_{}/{}"

url_query = {
    "order_currency": "BTC",
    "payment_currency": "KRW",
    "chart_intervals": "3m" 
}

# order_currency = "BTC"
# payment_currency = "KRW"
# chart_intervals = "3m"

# print(f'url: {url}')
# url = url.format()
# url = url.format(order_currency, payment_currency, chart_intervals)
url = url.format(**url_query)
print(f'url: {url}')

response = requests.request("GET", url)

print(response.text)
```
